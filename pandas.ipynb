<h3>Import pandas and read CSV file</h3>

```python
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
pd.set_option("display.max_rows", 10)

df1 = pd.read_csv("datset1.csv")
df2 = pd.read_csv("datset2.csv")
df1[:]
```

<h3>Dataset restructuring</h3>

```python
# Drop unnecessary columns
unnecessary_keys = set(df1.keys()).difference(list(df1.keys())[0:11])
df1.drop(unnecessary_keys, inplace=True, axis=1)
```

```python
# Drop unnecessary columns
unnecessary_keys = set(df2.keys()).difference(list(df2.keys())[0:11])
df2.drop(unnecessary_keys, inplace=True, axis=1)
```

```python
# Rename columns and values
new_keys = [
    'hero', 'rank', 'kda_ratio',
    'pick%', 'win%', 'elim_10min',
    'kill_10min', 'time_10min', 'dmg_10min',
    'heal_10min', 'dead_10min'
]

mapping = {}
i = 0
for k in df1.keys():
    mapping[k] = new_keys[i]
    i+=1

df1.rename(columns=mapping, inplace=True)
df1['hero'] = df1['hero'].str.lower()
df1['rank'] = df1['rank'].str.lower()

df2.rename(columns=mapping, inplace=True)
df2['hero'] = df1['hero'].str.lower()
df2['rank'] = df1['rank'].str.lower()
```

```python
# Fill null values
# Count null/missing values per column
df1.isnull().sum()
df1.fillna(0)
```

```python
# Melt ddataset
pd.melt(df1, id_vars=['hero'], value_vars=['pick%', 'win%'])
```

```python
# Imputing Data  with SimpleImputer
mod_df = df1.drop(['rank', 'hero'], axis=1)
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

df_imp = pd.DataFrame(imp_mean.fit_transform(mod_df), columns=mod_df.columns)
print(df_imp)
```

```python
# Concatenating 2 datasets  (Inner join)
pd.concat([df1, df2], ignore_index=True, join='inner')
```

```python
# Merging 2 datasets (Cartesian cross product)
df1.merge(df2, how='cross')
```
